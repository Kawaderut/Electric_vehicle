##### 12. Distribution of electric range among different vehicle models (using seaborn)

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Model', y='Electric Range', data=df)
plt.xticks(rotation=45)
plt.title('Distribution of Electric Range Among Different Vehicle Models')
plt.xlabel('Vehicle Model')
plt.ylabel('Electric Range')
plt.show()
```

#### 13. Is there a relationship between the electric range and the base MSRP?

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Electric Range', y='Base MSRP', data=df)
plt.title('Electric Range vs Base MSRP')
plt.xlabel('Electric Range')
plt.ylabel('Base MSRP')
plt.show()
```

##### 19. Are there any outliers in the Base MSRP column? If so, how many and which models?

```python
from scipy.stats import zscore

df['MSRP_zscore'] = zscore(df['Base MSRP'])
outliers = df[df['MSRP_zscore'] > 3]
num_outliers = len(outliers)
outlier_models = outliers['Model'].unique()
```
